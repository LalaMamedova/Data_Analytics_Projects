```python
import pandas as pd
```

```python
da_df = pd.read_excel("Data vacancy.xlsx")
ds_df = pd.read_excel("Data vacancy.xlsx",sheet_name="DS")
da_df.head(5)
```

```python
def separate_skills(data):
    df_copy = data.copy()
    skills_df = df_copy['Skills'].str.strip('[]').str.split(',').explode().str.strip().reset_index(drop=True)
    
    id_list = []

    for idx, row in df_copy.iterrows():
        skills = row['Skills'].strip('[]').split(',')
        skills = [skill.strip() for skill in skills] 
        id_list.extend([row['Id']] * len(skills))

    id_df = pd.DataFrame({'Id': id_list})
    
    return pd.concat([id_df, skills_df.reset_index(drop=True)], axis=1)

separated_da_skills = separate_skills(da_df)
separated_ds_skills = separate_skills(ds_df)
```

```python
separated_da_skills.head(15)
```

```python
with pd.ExcelWriter("Data vacancy.xlsx") as writer:
    da_df.to_excel(writer,sheet_name='DA',index=False)
    ds_df.to_excel(writer,sheet_name='DS',index=False)
    
    separated_da_skills.to_excel(writer,sheet_name='DA skills',index=False)
    separated_ds_skills.to_excel(writer,sheet_name='DS skills',index=False)
```
